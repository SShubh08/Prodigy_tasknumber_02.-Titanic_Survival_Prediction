```python
df['Age'].fillna((df['Age'].mean()),inplace=True)
```

```python
df =df.drop(labels=['Deck','Embark_town','Alive','Class','Alone','Adult_male','Who'],axis=1)
```

```python
df =df.dropna(axis=0)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
print(df['Sex'].unique())
print(df['Embarked'].unique())
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['Sex']=le.fit_transform(df['Sex'])
```

```python
df.head()
```

```python
df['Embarked']=le.fit_transform(df['Embarked'])
```

```python
df.head()
```

```python
df['Age'].hist(color='indigo',edgecolor='black')
plt.gca().set_facecolor('darkgrey')
plt.subplot(1, 2, 2)
sex_counts = df['Sex'].value_counts()
labels = sex_counts.index
colors = ['grey', 'brown'] 
plt.pie(sex_counts, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Sex Distribution')
plt.tight_layout()

plt.show()
```

```python
df['Age'].hist(color='indigo',edgecolor='black', grid=False)
plt.gca().set_facecolor('darkgrey')
plt.title('Histogram of Age', fontsize=16)
plt.xlabel('Age', fontsize=14)
```

```python
sns.boxplot(df['Age'],    
            color='darkgrey',  
            width=0.4,         
            fliersize=5, 
            linewidth=1.5,     
            notch=False)   
sns.set(style="dark")
plt.title('Boxplot of Age', fontsize=16)
plt.xlabel('Frequency', fontsize=14)
plt.ylabel('Age', fontsize=14)
plt.show()
```
